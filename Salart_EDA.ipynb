```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"F:\IVY\PYTHON\data\Salaries.csv")
```

```python
# Top 10 rows
df.head(10)
```

```python
# no of rows and columns
print("Number of rows",df.shape[0])
print("Number of columns",df.shape[1])
```

```python
df.info()
```

```python
# Replacing Not provided string with NAN 
for i in ["EmployeeName", "BasePay", "OvertimePay", "OtherPay","Benefits"]:
    df[i] = df[i].replace(["Not provided","Not Provided"],np.nan)
```

```python
for i in ["BasePay","OvertimePay","OtherPay","Benefits"]:
    df[i] = df[i].astype(float)
```

```python
df.isnull().sum()
```

```python
df.describe(include="all")
```

```python
# Dropping ID, Notes , Agency and Status as notes and Status is empty and Id and Agency is redundant
df.drop(columns=["Id","Notes","Agency","Status"],inplace=True)
```

```python
df.describe(include="all")
```

```python
# Top 5 employees by occurance 
df["EmployeeName"].value_counts().head(5)
```

```python
# No of unique job titles
print("There are",df["JobTitle"].nunique(), "no of unique job titles")
```

```python
# No of captians
print("There are",len(df[df["JobTitle"].str.contains("CAPTAIN", case=False)]),"captains")
```

```python
# No of employees working in fire department
print("There are",len(df[df["JobTitle"].str.contains("fire",case=False)]),"Employees working in fire department")
```

```python
# Max Min and avg pay
print(f"The max pay is {round(df["TotalPay"].max())}")
print(f"The avg pay is {round(df["TotalPay"].mean())}")
print(f"The minumum pay is {round(df["TotalPay"].min())}")
```

```python
# Rows having missing values 5 or more
df.drop(df[df.isnull().sum(axis=1)>=5].index, axis=0, inplace=True)
```

```python
# Find Job Title of Albert Pardini
print("Job title of Albert Pardini is",df[df["EmployeeName"]=="ALBERT PARDINI"]["JobTitle"])
```

```python
#person with highest pay
df[df["BasePay"]==df["BasePay"].max()]["EmployeeName"]
```

```python
# Avg base pay of all employees per year
df.groupby("Year")["BasePay"].mean()
```

```python
#avg base pay of all ACCOUNTANTS
df[df["JobTitle"]=="ACCOUNTANT"].groupby("JobTitle")["BasePay"].mean()
```
