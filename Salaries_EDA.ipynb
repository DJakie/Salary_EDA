```python
def salary_data(sal):
    if sal ==">50K":
        return 1
    else:
        return 0
```

```python
df["income"] = df["income"].apply(lambda x : salary_data(x))
```

```python
df.head()
```

```python
ax = sns.countplot(x="income", data=df)

# Add counts on top of each bar
for container in ax.containers: #ax.containers returns the bar containers used in the plot.
    ax.bar_label(container)     #ax.bar_label() annotates each bar with its height (the count).

plt.title("Income Distribution with Counts")
plt.show()
```

```python
# Salaries sorted by wrokclass
df.groupby("workclass")["income"].mean().sort_values(ascending=False)
```

```python
#Males have higher salary
df.groupby("gender")["income"].mean().sort_values(ascending=False)
```
